```python
train=pd.read_csv('/content/train.csv')
test=pd.read_csv('/content/test.csv')
```

```python
#handling the missing values
data['Item_Weight'].fillna(data['Item_Weight'].mean(), inplace=True)
data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0], inplace=True)
data['Item_Outlet_Sales'].fillna(data['Item_Outlet_Sales'].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Item_Type_Combined'] = data['Item_Identifier'].apply(lambda x: x[0:2])
data['Item_Type_Combined'] = data['Item_Type_Combined'].map({'FD':'Food', 'NC':'Non-Consumable', 'DR':'Drinks'})
data['Outlet_Years'] = 2013 - data['Outlet_Establishment_Year']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Outlet'] = le.fit_transform(data['Outlet_Identifier'])
var_mod = ['Item_Fat_Content', 'Outlet_Location_Type', 'Outlet_Size', 'Item_Type_Combined']
for i in var_mod:
    data[i] = le.fit_transform(data[i])
```

```python
# Split the data back into train and test
train = data[:len(train)]
test = data[len(train):]

# Split the train data into features and target variable
X_train = train.drop(['Item_Outlet_Sales'], axis=1)
y_train = train['Item_Outlet_Sales']
```

```python
X_train = X_train.apply(pd.to_numeric, errors='coerce')
X_train.fillna(0, inplace=True)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
# Train-test split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_val)
rmse_lr = np.sqrt(mean_squared_error(y_val, y_pred_lr))
print('RMSE for Linear Regression:', rmse_lr)
```

```python
# Predict on test data using the best model (Linear Regression)
X_test = test.drop(['Item_Outlet_Sales'], axis=1)
X_test = X_test.apply(pd.to_numeric, errors='coerce')
X_test.fillna(0, inplace=True)
test_predictions_lr = lr.predict(X_test)
```

```python
# Create submission file for Linear Regression
submission_lr = pd.DataFrame({
    'Item_Identifier': test['Item_Identifier'],
    'Outlet_Identifier': test['Outlet_Identifier'],
    'Item_Outlet_Sales': test_predictions_lr
})

submission_lr.to_csv('submission_lr.csv', index=False)
```

```python
# Plot the distribution of actual sales and predicted sales
plt.figure(figsize=(14, 7))
sns.histplot(train['Item_Outlet_Sales'], color='blue', label='Actual Sales', kde=True, stat="density", linewidth=0)
sns.histplot(test_predictions_lr, color='red', label='Predicted Sales (Test)', kde=True, stat="density", linewidth=0)
plt.legend()
plt.title('Comparison of Actual Sales and Predicted Sales')
plt.xlabel('Item_Outlet_Sales')
plt.ylabel('Density')
plt.show()
```

```python
# Scatter plot of Item_Weight vs. Predicted Sales
plt.figure(figsize=(14, 7))
plt.scatter(X_test['Item_Weight'], test_predictions_lr, alpha=0.3)
plt.title('Item_Weight vs. Predicted Item_Outlet_Sales')
plt.xlabel('Item_Weight')
plt.ylabel('Predicted Item_Outlet_Sales')
plt.show()
```

```python
import seaborn as sns
# Scatter plot with regression line
plt.figure(figsize=(14, 7))
sns.regplot(x=X_test['Item_Weight'], y=test_predictions_lr, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Item_Weight vs. Predicted Item_Outlet_Sales with Linear Regression Trendline')
plt.xlabel('Item_Weight')
plt.ylabel('Predicted Item_Outlet_Sales')
plt.show()
```

```python
import statsmodels.api as sm

# Function to plot scatter plot with OLS trendline
def scatter_plot_with_ols(x, y, data, xlabel, ylabel, title):
    # Scatter plot
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x, y=y, data=data, alpha=0.5)

    # OLS trendline
    X = data[x]
    X = sm.add_constant(X)
    Y = data[y]
    model = sm.OLS(Y, X).fit()
    predictions = model.predict(X)

    # Plot the trendline
    plt.plot(data[x], predictions, color='red', label='OLS Trendline')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.show()

# Plot Item MRP vs Item Outlet Sales with OLS trendline
scatter_plot_with_ols('Item_MRP', 'Item_Outlet_Sales', train, 'Item MRP', 'Item Outlet Sales', 'Item MRP vs Item Outlet Sales')

# Plot Outlet Years vs Item Outlet Sales with OLS trendline
scatter_plot_with_ols('Outlet_Years', 'Item_Outlet_Sales', train, 'Outlet Years', 'Item Outlet Sales', 'Outlet Years vs Item Outlet Sales')

# Plot Item Weight vs Item Outlet Sales with OLS trendline
scatter_plot_with_ols('Item_Weight', 'Item_Outlet_Sales', train, 'Item Weight', 'Item Outlet Sales', 'Item Weight vs Item Outlet Sales')
```
